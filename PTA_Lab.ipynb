# Binomial

```python
# n=300 X=100
from scipy.stats import bernoulli
from scipy.stats import binom
import seaborn as sb

data_bern1 = binom.rvs(n=300,size=300,p=0.027)
ax = sb.histplot(data_bern1,
                  kde=True,
                  color='crimson',
                 # hist_kws={"linewidth": 25,'alpha':1}
                )
ax.set(xlabel='Binomial', ylabel='Frequency')
```

# Poisson

```python
# n=300 lambda= 8.1
from scipy.stats import poisson
# import seaborn as sb

data_pois = poisson.rvs(mu=8.1,size=300)
ax = sb.distplot(data_pois,
                  kde=True,
                  color='crimson',
                 # hist_kws={"linewidth": 25,'alpha':1}
                )
ax.set(xlabel='Poisson', ylabel='Frequency')
```

# Stacked Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 4,sharey=False, tight_layout=True)
fig.set_figheight(7)
fig.set_figwidth(15)
```

```python
axs[0].hist(data_bern1,bins=15,color = "red")
```

```python
axs[1].hist(data_bern2,bins=15,color = "yellow")
```

```python
axs[2].hist(data_bern3,bins=15,color = "green")
```

```python
axs[3].hist(data_pois,bins=15,color = "blue")
```
